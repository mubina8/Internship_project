```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("dataMarketing.csv")
df.rename(columns={"Campaign ID":"Campaign_ID","Campaign Name":"Campaign_Name","Unique Clicks":"Unique_Clicks",
        "Unique Link Clicks (ULC)":"ULC",
        "Click-Through Rate (CTR)":"CTR","Unique Click-Through Rate (Unique CTR)":"Unique_CTR",
       "Amount Spent in INR":"Amount_in_INR","Cost per Result (CPR)":"CPR","Cost Per Click (CPC)":"CPC"},inplace=True)
```

```python
df
```

```python
plt.title("Campaign Cost")
plt.ylabel("Campaign ID")
plt.xlabel("Total Cost($)")
df_sort = df.sort_values('Amount_in_INR')
plt.barh(df.Campaign_ID,df.Amount_in_INR,data=df_sort)
plt.rcParams["figure.figsize"] = (25,5)
plt.grid()
```

```python
plt.title("Campaign Cost",fontsize='30')
plt.ylabel("CTR(%)",fontsize='30')
plt.xlabel("Frequency",fontsize='30')
plt.scatter(df.Frequency,df.CTR,s=100,alpha=0.5,color='green')
plt.rcParams["figure.figsize"] = (25,5)
plt.grid()
plt.legend()
```

```python
plt.title("Campaign Click Through Rate",fontsize='20')
plt.ylabel("CTR",fontsize='20')
plt.xlabel("Campaign ID",fontsize='20')
plt.bar(df.Campaign_ID,df.CTR,color='green')
plt.rcParams["figure.figsize"] = (25,5)
plt.grid()
```

```python
plt.title("Campaign Cost per click",fontsize='20')
plt.ylabel("CPC",fontsize='20')
plt.xlabel("Campaign ID",fontsize='20')
plt.bar(df.Campaign_ID,df.CPC,color='Red')
plt.rcParams["figure.figsize"] = (25,5)
plt.grid()
```

```python
plt.title("Campaign Cost per Result",fontsize='20')
plt.ylabel("CPR",fontsize='20')
plt.xlabel("Campaign ID",fontsize='20')
plt.bar(df.Campaign_ID,df.CPR,color='Yellow')
plt.rcParams["figure.figsize"] = (25,5)
plt.grid()
```

```python
plt.title("Campaign Student Age distribution",fontsize='20')
plt.bar(df.Age,df.Reach)
plt.ylabel("Reach",fontsize='20')
plt.xlabel("Age",fontsize='20')
```
